```python
# импортируем файл с нашим ключом, чтобы ключ не был в открытом доступе
# import api_key
# api_key.your_client_access_token
your_client_access_token = "i cleaned it so for now :) "
```

```python
client_access_token = your_client_access_token
search_term = "MotherMother"
genius_search_url = f"http://api.genius.com/search?q={search_term}&access_token={client_access_token}"
print(genius_search_url)
```

```python
import requests
response = requests.get(genius_search_url)
json_data = response.json()
json_data
```

```python
for song in json_data['response']['hits']:
    print(song['result']['full_title'])
```

```python
import pandas as pd
from IPython.core.display import HTML

def get_image_html(link):
    image_html = f"<img src='{link}' width='50px'>"
    return image_html

mother_mother_songs = []
for song in json_data['response']['hits']:
    mother_mother_songs.append([song['result']['full_title'], song['result']['stats']['pageviews'], song['result']['song_art_image_url']])

mother_mother_df = pd.DataFrame(mother_mother_songs)
mother_mother_df.columns = ['song_title', 'page_views','album_cover_url']

#Use the function get_image_html()
mother_mother_df['album_cover'] = mother_mother_df['album_cover_url'].apply(get_image_html)

mother_mother_df
```

```python
HTML(mother_mother_df[['album_cover', 'page_views', 'song_title']].to_html(escape=False))
```
